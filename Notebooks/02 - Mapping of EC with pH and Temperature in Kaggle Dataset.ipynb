```python
import joblib
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../Data/Kaggle Dataset of N, P, K, pH, Temperature and Labels.csv')
df.head()
```

```python
X = df[['pH', 'Temperature']]
X.head()
```

```python
model_01 = joblib.load('../Models/EC - pH Adjusted Data Model.h5')
model_02 = joblib.load('../Models/EC - pH Original Data Model.h5')
EC_01 = model_01.predict(X)
EC_02 = model_02.predict(X)

print(
	model_01.predict(
		[
			[7.0, 25]
		]
	)
)
```

```python
pH = df['pH']
plt.scatter(pH, EC_01, label = "EC 01 - pH Relation")
plt.xlabel('pH')
plt.ylabel('EC 01')
plt.title('Comparison Of EC 01 & pH')
plt.legend()
plt.show()

pH = df['pH']
plt.scatter(pH, EC_02, label = "EC 02 - pH Relation")
plt.xlabel('pH')
plt.ylabel('EC 02')
plt.title('Comparison Of EC 02 & pH')
plt.legend()
plt.show()
```
